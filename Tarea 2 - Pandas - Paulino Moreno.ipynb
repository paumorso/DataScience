# Lo Básico

Ahora que tenemos nuestro set de datos en nuestra variable df, veamos qué contiene. Podemos usar la función **head()** para ver las primeras 5 filas (o la función **tail()** para ver las últimas 5 filas).

```
df.head()
```

```
df.tail()
```

Podemos ver las dimensiones del set de datos utilizando el atributo o propiedad **shape**.

```
df.shape
```

También podemos extraer todos los nombres de columna como una lista, usando el atributo **columns** y puede extraer las filas con el atributo **index**.

```
df.columns.tolist()
```

```
df.index.tolist()
```

Para tener una mejor idea del tipo de datos con los que estamos tratando, podemos llamar a la función **describe ()** para ver estadísticas como media, min, etc. sobre cada columna del conjunto de datos.

```
df.describe()
```

Bien, ahora veamos la información que queremos extraer del set de datos. Digamos que quería saber el valor máximo de una determinada columna. La función **max ()** le mostrará los valores máximos de todas las columnas

```
df.max()
```

Luego, si desea obtener específicamente el valor máximo para una columna en particular, use el nombre de la columna entre comillas dobles dentro de paréntesis cuadrados en la variable df (que contiene el set de datos), de la siguiente manera:

```
df['WScore'].max()
```

Si desea encontrar la media de la puntuación de los equipos perdedores.

```
df['LScore'].mean()
```

Si queremos ver realmente el juego (fila) donde ocurrió este puntaje máximo. Podemos llamar a la función **idxmax ()** para identificar el índice de fila

```
df['WScore'].idxmax()
```

Una de las funciones más útiles que puede invocar en ciertas columnas en un marco de datos es la función **value_counts ()**. Muestra cuántas veces aparece cada elemento en la columna. Este comando en particular muestra el número de juegos en cada temporada

```
df['Season'].value_counts()
```

# Filtrando filas condicionalmente

Ahora, digamos que queremos encontrar todas las filas que satisfacen una condición particular. Por ejemplo, quiero encontrar todos los juegos donde el equipo ganador anotó más de 150 puntos. La idea detrás de este comando es que desea acceder a la columna 'WScore' del set de datos df (df ['WScore']), buscar qué datos están por encima de 150 (df ['WScore']> 150), y luego devuelve solo aquellas filas específicas en un formato de set de datos (df [df ['WScore']> 150]).

```
df[df['WScore'] > 150]
```

Esto también funciona si tiene múltiples condiciones. Digamos que queremos saber cuándo el equipo ganador anota más de 150 puntos y cuándo el equipo perdedor anota menos de 100.

```
df[(df['WScore'] > 150) & (df['LScore'] < 100)]
```
