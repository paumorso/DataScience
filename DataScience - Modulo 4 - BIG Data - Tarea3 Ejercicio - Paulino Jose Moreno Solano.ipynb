### Parte 2

Continuamos con la carga en un conjunto de datos de Spark los datos de las transacciones del conjuto de datos de las Compras, Fechas, Productos, Agencias, Clientes y Rutas, para que estas sean ingresadas en la Tabla de PostgreSQL.

Para lo cual debemos crear la estrutura de la tabla en PostgreSQL:

DROP TABLE IF EXISTS "bimbo"

CREATE TABLE "public"."bimbo"
(
    "id"                 integer                        NOT NULL,
    "Semana"             integer                        NOT NULL,
    "Agencia_ID"         integer                        NOT NULL,
    "Canal_ID"           integer                        NOT NULL,
    "Ruta_SAK"           integer                        NOT NULL,
    "Cliente_ID"         integer                        NOT NULL,
    "Producto_ID"        integer                        NOT NULL,
    "Fecha"              character varying(10)          NOT NULL,
    "NombreProducto"     character varying(500)         NOT NULL,
    "Ciudad"             character varying(500)         NOT NULL,
    "Estado"             character varying(500)         NOT NULL,
    "NombreCliente"      character varying(500)         NOT NULL,
    "Ruta_Ciudad"        character varying(500)         NOT NULL,
    "Ruta_Estado"        character varying(500)         NOT NULL
) WITH (oids = false)

Iniciamos con la carga de las librerias para de PySpark.

```python
# Importar librerias.
import findspark
findspark.init('C:\Spark\spark-2.4.4-bin-hadoop2.7')

from datetime import datetime
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, date_format, udf 
from pyspark.sql.types import DateType

from pyspark.sql.functions import to_timestamp
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

Configuramos la sesión de Spark.

```python
# Crear sesión de Spark
spark = SparkSession \
    .builder \
    .appName("pyspark_bimbo_app") \
    .config("spark.driver.extraClassPath", "postgresql-42.1.4.jar") \
    .config("spark.executor.extraClassPath", "postgresql-42.1.4.jar") \
    .master("local[*]") \
    .getOrCreate()
```

Guargamos los datos en una estructura básica para ser usados en la conexión con PostgreSQL.

```python
# Registro de datos de credenciales
properties = {
    'driver': 'org.postgresql.Driver',
    'url': 'jdbc:postgresql://localhost/postgres',
    'user': 'postgres',
    'password': 'paumorso',
    'dbtable': 'bimbo',
}
```

Conexión y lectura de la tabla de PostgreSQL donde se nota que no hay datos todavía cargados.

```python
# Configurar conexión para lectura de tabla en PostgreSQL.
Spark_Bimbo_Lectura = spark.read \
    .format('jdbc') \
    .option('driver', properties['driver']) \
    .option('url', properties['url']) \
    .option('user', properties['user']) \
    .option('password', properties['password']) \
    .option('dbtable', properties['dbtable']) \
    .load()

# Prueba de carga de datos.
Spark_Bimbo_Lectura.show()
```

Configuración de la estructura de Spark para cargar el conjunto de datos.

```python
# Configurar el DataFrame en Spark.
Spark_Bimbo_Schema = StructType([
    StructField('id', IntegerType(), True),
    StructField('Semana', IntegerType(), True),
    StructField('Agencia_ID', IntegerType(), True),
    StructField('Canal_ID', IntegerType(), True),
    StructField('Ruta_SAK', IntegerType(), True),
    StructField('Cliente_ID', IntegerType(), True),
    StructField('Producto_ID', IntegerType(), True),
    StructField('Fecha', StringType(), True),
    StructField('NombreProducto', StringType(), True),
    StructField('Ciudad', StringType(), True),
    StructField('Estado', StringType(), True),
    StructField('NombreCliente', StringType(), True),
    StructField('Ruta_Ciudad', StringType(), True),
    StructField('Ruta_Estado', StringType(), True)
])

#Estructura de la tabla (columnas del conjunto de datos).
#id,Semana,Agencia_ID,Canal_ID,Ruta_SAK,Cliente_ID,Producto_ID,Fecha,NombreProducto,Ciudad,
#Estado,NombreCliente,Ruta_Ciudad,Ruta_Estado
```

Cargamos el conjuto de datos de las Compras, Fechas, Productos, Agencias y Clientes con el conjunto de datos de las Rutas en la estructura de Spark para ser cargada a PostgreSQL.

```python
# Cargar DataFrame de Spark con los datos.
Spark_Bimbo = spark.createDataFrame(dsCompraFechaProductoAgenciaClienteRuta, Spark_Bimbo_Schema)

# Prueba de carga de datos.
Spark_Bimbo.show()
```

Conexión y escritura de la tabla de PostgreSQL de los datos que están en la estructura de Spark.

```python
# Configurar conexión para escritura de tabla en PostgreSQL.
Spark_Bimbo.write \
    .format('jdbc') \
    .option('driver', properties['driver']) \
    .option('url', properties['url']) \
    .option('user', properties['user']) \
    .option('password', properties['password']) \
    .option('dbtable', properties['dbtable']) \
    .mode('append') \
    .save()
```

Conexión y lectura de la tabla de PostgreSQL ya con los datos cargados.

```python
Spark_Bimbo_Lectura = spark.read \
    .format('jdbc') \
    .option('driver', properties['driver']) \
    .option('url', properties['url']) \
    .option('user', properties['user']) \
    .option('password', properties['password']) \
    .option('dbtable', properties['dbtable']) \
    .load()

# Prueba de carga de datos.
Spark_Bimbo_Lectura.show()
```
