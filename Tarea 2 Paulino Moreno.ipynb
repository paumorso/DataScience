### Introducción

### Visualización con Matplotlib

Lo siguiente importa el módulo principal en el paquete `matplotlib`. Tenga en cuenta que `%matplotlib inline` significa que queremos que la gráfico está en línea en nuestro notebook en lugar de obtener una ventana emergente.

````
from matplotlib import pyplot as plt
%matplotlib inline
````

La visualización es una de las cosas más importantes del análisis de datos. Además de producir gráficos legibles, debemos hacer un esfuerzo para mejorar el atractivo general de las gráficos. matplotlib es un poderoso paquete para usuarios de ___Python___. Vamos a empezar con un ejemplo.

Primero generamos datos de muestra.

````
X_arr = np.arange(10)
Y_arr = 3*X_arr + 2 + np.random.random(size=X_arr.size)
print(X_arr)
print(Y_arr)
````

Para trazar un diagrama de dispersión simple, podemos usar la función `plt.scatter()`

````
plt.scatter(X_arr, Y_arr)
````

####  Diferentes gráficos

Además de los diagramas de dispersión / línea más simples, existen otras opciones.

##### Histogramas

Usemos una distribución gaussiana como ejemplo.

````
mu, sigma = 15, 1
gaussian_arr = np.random.normal(mu, sigma, size=10000)
np.mean(gaussian_arr), np.std(gaussian_arr, ddof=1)
````

````
fig, ax = plt.subplots()
# `hist()` Devolveremos algo pero por lo general no necesitamos.
freq_arr, bin_arr, _ = ax.hist(gaussian_arr)
````

Podemos personalizarlo y hacerlo más bonito.

````
fig, ax = plt.subplots()
# Facecolor establecido en verde; transparencia (`alpha`) nivel: 30%
freq_arr, bin_arr, _ = ax.hist(gaussian_arr, facecolor='g', alpha=0.3)
# Añadir rejilla
ax.grid()
````

````
mu, sigma = 15, 1
gaussian_arr = np.random.normal(mu, sigma, size=10000)
np.mean(gaussian_arr), np.std(gaussian_arr, ddof=1)
````

````
fig, ax = plt.subplots()
# `hist()` will return something but we usually do not need.
freq_arr, bin_arr, _ = ax.hist(gaussian_arr)
````

##### Boxplot

Con los datos de la guaussiana, podemos usar una forma diferente: box plot

````
fig, ax = plt.subplots()
ax.boxplot(gaussian_arr, 
           vert=True,  # vertical
           showfliers=False, # no mostrar valores atípicos (outliers)
           showmeans=True, # mostrar la media
           labels=['Gaussian'] # nombre del grupo (etiquetas)
          )
````

##### Gráfico de barras

Datos aleatorios para la entrada del gráfico de barras

````
bar_arr = np.array(['Primavera', 'Verano', 'Otoño', 'Invierno'])
freq_arr = np.random.randint(0, 100, 4) # generar 4 enteros aleatorios para el gráfico de barras
freq_arr
````

También podemos añadir barras de error:

````
yerr_arr = np.random.randint(5, 10, 4)
yerr_arr
````

````
fig, ax = plt.subplots()
ax.bar(bar_arr, freq_arr, # X , Y
       yerr = yerr_arr, # barras de error
       color='gold'
      )
````

##### Agrupar estos gráficos

````
fig, ax_arr = plt.subplots(nrows=2, ncols=2, 
                           # que sus escalas sean diferentes
                           sharex=False, sharey=False)
fig.set_figwidth(12)
fig.set_figheight(8)
# establecer título global
fig.suptitle("Mis primeros gráficos")

## 1er
ax_arr[0, 0].scatter(X_arr, Y_arr)
ax_arr[0, 0].axis([0, 10, 0, 35])
ax_arr[0, 0].set_title('Mi 1er Gráfico')
## 2do 
ax_arr[0, 1].hist(gaussian_arr, facecolor='g', alpha=0.3)
ax_arr[0, 1].set_title('Histograma')
ax_arr[0, 1].grid()
## 3ero
ax_arr[1, 0].boxplot(gaussian_arr, vert=True, showfliers=False,
                     showmeans=True, labels=['Gaussian'])
ax_arr[1, 0].set_title('Box plot')
## 4to
ax_arr[1,1].bar(bar_arr, freq_arr, 
                yerr = yerr_arr, color='gold')
ax_arr[1, 1].set_title('Gráfico Barras')
````
